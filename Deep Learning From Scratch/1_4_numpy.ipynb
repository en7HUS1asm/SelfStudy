```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y1, label = "sin")
plt.plot(x, y2, linestyle="--",color="red", label = "cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sin & Cos")
plt.legend()
plt.show()
```
